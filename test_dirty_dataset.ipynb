```python
# --------------------------------------------------
# Negative Values
# --------------------------------------------------
numeric_cols = df.select_dtypes(
    include=["int64", "float64"]
).columns

negative_count = (
    (df[numeric_cols] < 0)
    .sum()
    .sum()
)

print("\nNegative Values:",
      negative_count)
```

```python
# --------------------------------------------------
# Blank Strings
# --------------------------------------------------
df = pd.read_csv(
    "test_both_clean_n_dirty_dataset_CLEAN_SOLUTION.csv",
    keep_default_na=False
)

blank_count = (
    (df == "")
    .sum()
    .sum()
)

print("Blank Values:", blank_count)
```

```python
# --------------------------------------------------
# Outliers (IQR Method)
# --------------------------------------------------
print("\nOUTLIERS")

# for col in numeric_cols:

#     q1 = df[col].quantile(0.25)
#     q3 = df[col].quantile(0.75)

#     iqr = q3 - q1

#     lower = q1 - 1.5 * iqr
#     upper = q3 + 1.5 * iqr

#     outliers = (
#         (df[col] < lower) |
#         (df[col] > upper)
#     ).sum()

#     if outliers > 0:
#         print(f"{col}: {outliers}")

# print("\nCHECK COMPLETED")

for col in numeric_cols:

    temp = pd.to_numeric(
        df[col],
        errors="coerce"
    )

    q1 = temp.quantile(0.25)
    q3 = temp.quantile(0.75)

    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outliers = temp[
        (temp < lower) |
        (temp > upper)
    ]

    print(f"{col}: {len(outliers)}")
```
